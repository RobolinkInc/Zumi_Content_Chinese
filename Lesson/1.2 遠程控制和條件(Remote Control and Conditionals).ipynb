## If statements（條件語句）
<font size =3>在偽代碼中，您的編程將根據您的輸入來須指示Zumi朝特定方向行駛。條件語句（If statements）, or **有條件的（conditionals）**, 是用作幫助您的編程去做出決定。其實我們每天也會使用條件語句！以下是一些例子:


* 如果我的鬧鐘響了，那我會醒來。否則，我會保持睡眠狀態。
* 如果我餓了，我會吃點東西。
* 如果晴天，我會戴防曬霜。
* 如果我的房間乾淨，我將得到津貼。 否則，我不會賺錢。

條件語句在Python的寫法: <br> <br>

<font face="Courier">
    
if insert condition here: <br>
<span style="margin-left: 40px;">Insert statements here</span> <br>
else: <br>
<span style="margin-left: 40px;">Insert statements here</span> <br>

if alarm goes off:<br>
<span style="margin-left: 40px;">I will wake up</span><br>
else:<br>
<span style="margin-left: 40px;">Stay asleep</span><br>
</font>

<img src="../Data/images/conditional_flowchart.jpg" width=700>



條件是計算機(computer)用作檢查它是對或錯的指標。
* 如果條件是對的，那計算機將執行或執行其下的語句。
* 如果條件是錯的，那計算機將不會執行其下面的語句。
* "否則（else)"是沒有條件的! 因為"否則（else)" 是使用if-else語句時的最後一個選項。當計算機在檢查了所有其他條件後到達"否則(else)"選項時，"否則(else)"條件將自動為真，並運行其語句。
*其中一種使用“否則(else)”的方法是當條件被否定時需要發生某些事情。當“if(如果)”條件為錯/假，並且不會做出任何反應時，請不要使用“否則(else)”。
### Indentation(縮排)

縮排在Python編程語言中非常重要。 縮排指示哪些語句屬於代碼中的特定部分。

您可能已經註意到，if和else語句下方的語句縮進了4個空格或一個製表符空間(tab space)。即使縮排只是相差一個空白鍵的距離，您也會收到錯誤訊息！嘗試單一使用製表符(tab)或空格(space bar)，不要將兩者結合在一起。

## Comparators（比較器）
<font size =3>接下來，將您的偽代碼轉換為Python代碼。 在輸入驅動器命令之前，請使用打印語句(printe statement)進行測試，如下所示： <br> <br>
    
<font face="Courier"> 
direction = input("Please enter a command: ")

if direction == "w": <br>
<span style="margin-left: 40px;">print("Go forward")</span> <br>    
</font>
    
等號可能看起來很熟悉，但是在代碼中它的應用略有不同！ <font face="Courier">==</font> (等號) **(比較器)comparator** 的一個例子。比較器(Comparators)可以在內部條件下用於比較值。如果該語句為對(true)，則將執行其下面的語句。這些都是比較器：

* \> Greater Than
* \< Less Than
* \>= Greater Than or Equal To
* <= Less Than or Equal To
* == Equal To
* != Not Equal To 

運行下面的單元格，並對條件進行一些更改，以了解它們的工作原理！ 如果需要，可以更改變量名稱或輸入語句。請確保測試您的代碼沒有錯誤並能正常運作。
</font>


```python
X= 10
Y = 100
Z = 200

if X != Y:
    print("X is not equal to Y")

if Z > Y:
    print("Z is greater than Y")
```
