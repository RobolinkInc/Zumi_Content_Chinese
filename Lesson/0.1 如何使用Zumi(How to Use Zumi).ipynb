# What is Zumi?(什麼是Zumi)
<font size =3>Zumi是第一部學術性的人工智能自動駕駛汽車，它會以有趣並生動的方式教導大家人工智能和自動駕駛汽車技術的概念。除了AI外，您亦會學到一些基本的機械人及編程的概念！</font>

# What is AI?（什麼是AI？)
 <font size =3>AI的全寫是artificial intelligence，即是人工智能的意思。聽起來可能有些嚇人，但其實您每天都在與人工智能互動！例如:當Netflix向您提出建議或您在詢問Siri或Alexa要到達您最喜歡的餐廳需要多長時間時，其實您都正在與人工智能進行互動。 人工智能軟件設計得像人一樣聰明，並且能做到以下事情：

* Natural language processing（自然語言處理(NLP)）。其中一個例子是任何文本都具有文字轉語音功能(text-to-speech feature)，即大聲朗讀螢幕上的文本。對任何有視力障礙的人來說都是很棒的功能！
* Object classification(物件分類)。這是圖像被歸類為正確物件的時候。一些AI能做到更深入，什至能對花的種類或狗的品種進行分類。
* Reasoning and problem-solving(推理及解決問題)。有些AI是設計來玩遊戲，例如Tic-Tac-Toe(過三關)及Battleship(海戰棋)，利用「條件」來決定它們的下一步棋。例如: 它的邏輯可能是這樣的，"如果我的對手在加上角放了一個「X」，我便會在它的左面放「O」。否則，我就會在右上角放「O」"。

AI有多種形式，它們並不是經常以機械人的形式出現。您即將踏上學習如何將AI應用在自動駕駛汽車中的旅程！</font>

## General Precautions(一般注意事項)

* <font size =3>避免在手濕的時候使用Zumi亦應遠離液體使用。</font>
* <font size =3>請勿將金屬或導電物料插入Zumi的外殼或電路版附近。</font> 
* <font size =3>請在開機前確保所有與Zumi的連接均正確無誤。</font>
* <font size =3>在扭緊螺栓或支座時注意不要夾到任何電線。</font>
* <font size =3>在組合Zumi或接駁其他配件時請先把Zumi關機。</font>


<a id='battery'></a>
## Battery Specifications and Usage(電池規格和使用)
<font size =3>Zumi只能以鋰離子聚合物電池(Lithium Polymer battery)供電(3.7V-4.2V)。該電池有一個標準的JST PH 2.0MM connector(接駁口)。</font>
#### Connecting the battery(連接電池)
<font size =3>在連接電池時，請確保紅線電線接駁到PCB(電路版)的(+)，而黑線電線則接駁到PCB的(-)。請**不要**把電池接駁到電路版上的其他接駁口上，這會導致電池和Zumi電路版損壞。</font>

#### Removing the battery(移除電池)
<font size =3>如果您需要移除電池，抓住電路接駁器上的白色塑料接口。一把鉗子就可以完成這項工作。避免從電線上拉扯電池。</font>

#### Low battery(低電量)
<font size =3>當電池電量變低時，Zumi會發出嗶嗶聲。如果您手上沒有MicroUSB，請先把Zumi關上，不這樣做的話Zumi的功率輸出會隨著時間有非常大的下降。如果您手上有MicroUSB請閱覽「如何為Zumi充電」。</font>

<a id='charging'></a>
## Charging Zumi(為Zumi充電)
<font size =3>Zumi電路版的左面有3盞LED，它們會表示現時的充電狀態</font>

* <font size =3>如果紅色LED有亮起，代表電池正在充電。.</font>
* <font size =3>如果藍色LED有亮起，代表檢測到MicroUSB的電源。.</font>
* <font size =3>如果綠色LED有亮起，代表Zumi正在通電。</font>
* <font size =3>如果 **只有** 綠色LED有亮燈，代表Zumi正使用電池的電力。</font>

#### How to charge Zumi(如何為Zumi充電？)
<font size =3>Zumi只會在紅色LED有亮卷及MicroUSB接駁到電源(電腦USB端口，插座上的USB接口...)時進行充電。如果您接駁MicroUSB後發現紅色LED**沒有**亮起，請您把開關(switch)推至開啟的狀態以進行充電。</font>

#### How do I know if Zumi is charged?(我們如何知道Zumi已完成充電？)
<font size =3>如果開關處於充電位置（“關閉”位置）且紅色LED沒亮，那就電池已完成充電。</font>

#### Can I use Zumi while charging?(我能在Zumi充電使用它嗎？)
<font size =3>您能在Zumi充電時使用它，但有些駕駛指令可能會因接駛著MicroUSB cable而受限制。如果您在充電時使用Zumi並想把USB cable斷開，您**必需**把開關撥到"開啟位置"。Zumi將會發出嗶一聲，紅色LED會隨即關閉，而綠色跟藍色LED會亮起。在此刻您可能會斷開連接。如果您在撥開關前Zumi已經斷開連接的話，Zumi會關機，您需要重新啟動它。</font>

<a id='zumiboard'></a>
## Zumi Board Components(Zumi電路板組件)

<font size =3>下面列出了一些連接到PCB(電路版)或PCB上的配件的注意事項。</font>

#### Infrared (IR) LEDs(紅外線LED)
<font size =3>IR LEDs(特別是底部跟前方的LEDs)都有很長的電線，非常容易彎曲，特別是在建造Zumi時。在開啟Zumi前，請確保腳位(PINS)電線沒有被壓住。</font>

#### Motors(馬達)
<font size =3>要避免Zumi的馬達受損，請好好對待它的小齒輪箱(gearbox)。馬達有一個非常強力的減速齒輪箱，這使它要變得非常細小同時亦非常易損壞。不要強逼馬達這樣轉動，這會剝落齒輪的齒並導致這馬達無法再使用。避免在馬達轉動時緊握它以讓馬達停下。如果您的Zumi在關上馬達時出現問題的話，可能是因為馬達在消耗電力時太快，您可能需要降低輸入馬運轉動的速度。
</font>

#### Arduino reset button(Arduino 重置接鈕)
<font size =3>如果Zumi不會在運行特定代碼後停止運作或對指令沒有任何反應(馬達持續失控)，您可以按下Zumi左側的reset button來重置Arduino電路版。您可以通過從內部的彈出窗口來訪問重置按鈕。請記住，這只會重置Arduino電路板，而不是Raspberry Pi。如果 Pi出現問題，重置按鈕是起不到作用的。</font>
    
#### Camera(相機)
<font size =3>相機是非常昂貴的，所以請您在任何時間都為Zumi裝上外殼！如果您在執行編程時得到一個資源(resources)出現**MMAL**錯誤，您的相機可能會失去連接。檢查相機前面的細小接駁器，及連接到Pi的連接器。如果您需要一些視覺效果，請查看我們的 FAQ :)
    

<a id='poweroff'></a>
## Powering off(關機)
<font size =3>Zumi是一部小電腦並且需要正確地關機。
請利用dashboard(儀表版)來關閉Zumi，在10秒後再利用開關(switch)關閉電源。
如果您在使用終端(terminal),使用```sudo poweroff```指令,等10秒後再將開關撥至"閉機位置"。</font>
