## Cells(單元格)

<font size =3>一個典型的Jupyter notebook會劃分為幾個單元格。一些單元格有文字，例如您正在閱覽的這個，一些能夠運行程式。</font>

### Running code(運行編程)
<font size =3>以下的單元格是一個代碼單元格的例子。要運行單元格並執行編程/代碼，請選擇高亮為<span style="color:dodgerblue">藍色</span>或<span style="color:#52BE80">綠色</span>。*以供將來參考，藍色的單元格概述(cell outline)被選擇但它未處於編輯模式。綠色的單元格概述被選下並且處於編輯模式。在任何狀態下，您都能在toolbad(工具欄)點擊在"RUN"或按下<font face="Courier">Shift</font> + <font face="Courier">Enter</font>。</font>*

```python
print("This is a code cell!")
```

<font size =3> 在您運行單元格後，編程的輸出將會在下面呈現出來。單元格旁邊的數字亦會由 <font face="Courier">In[]</font>轉換成<font face="Courier">In[1]</font>。那個數字代表單元格完成或執行了多少次指令。在Jupyter Notebook,您能一次或分開幾部分來運行整個編程！嘗試重覆運行上面的編程，<font face="Courier">In[1]</font>將會變成 <font face="Courier">In[2]</font>。

如果<font face="Courier">In[]</font>指示符沒有在單元格旁邊出現，它會**不能**運行編程的。例如，"運行"以下的單元格。</font>


print("This is not a code cell")

<font size =3>這個單元格被格式化為文本。這些單元格都稱為**markdown**單元格。有時候您可能會把您的單元格不小心轉換成不同類型。如果發生這些意外，在toobar上選擇"Cell"然後再選"Cell Type"以改回您所需的類型。嘗試把上面的**markdown**單元格改成**code(編程)**。</font>


### Stopping the code(停止代碼)
<font size =3> 有時您需要停止當前正在運行的代碼。可能是您的while loop正在無限輪迥，您需要擺脫它！工具欄上"Run"旁邊您會找到一個黑方正方形。這是一個停止按鈕當按下它單元格正在運行的程式便會立即中止。運行以下的單元格(它會不斷運行)然後停止這個代碼。</font>

```python
count=0
while True:
    print(count)
    count+=1
    
```

<font size =3> 您在單代格運行時有發現旁邊有一個**asterisk(星號)**嗎？看起來像這樣：<font face="Courier">In[*]</font>。有星號在旁的單元格代表它正在運行。請避免在單元格正在運行時再運行更多個單元格。</font>

## Notebook functions(筆記本的功能)

### Restarting a notebook(重新開啟筆記本)
<font size =3>您可能會想清除目前單元格中所有的輸出並重設所有的編程。在工具欄中有一個叫作"Kernel"的欄標，在它的清單中您能看見一個功能叫作"Restart & Clear Output"。當您選擇它後，所有的輸出會消除，單元格亦能再次運行。您再次檢查<font face="Courier">In[]</font>是否已清空來確保它已被重設。 brackets. </font>
