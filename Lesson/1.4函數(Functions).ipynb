# Creating Custom Functions(創建自定義函數)

<font size =3> 函數是通過將常用的代碼塊存儲至一個命名集中來節省時間和化簡編程的方法。 你在呼叫/使用<font face="Courier">zumi.forward()</font> or <font face="Courier">zumi.reverse()</font>的時候其實已經在使用函數了。
但其實您也能創建屬您自己的函數！
</font>

### Defining functions(定義函數)
<font size=3> 編寫自己的函數的第一步是關鍵字<font face="Courier">「def」</font>. 您將需要令Python明白您正在**define(定義)**一個函數。接下來，利用括號跟冒號來命名您的函數。在冒號之後，縮排並放進函數的主體，主體即是函數接一下將會執行的操作。以下是一個簡單的例子：</font>

```python
def greet_user():
    print("Hello")
```

<font size=3>如果您運行上面的代碼，任何事情都不會發生。這組代碼只是在定義函數(great_user)，您需要呼叫(call)這個函數才會運行。 </font>

```python
greet_user() # Call the function
```

### Defining functions with parameters(利用變數來定義函數)

*   清單項目
*   清單項目


<font size=3> 如果我們希望函數根據所提供的參數而執行不同的操作該怎麼辦？還記得我們可以透過改變<font face="Courier">zumi.forward()</font>的秒數嗎?同樣地我們亦能將參數加至函數中！</font>

```python
def greet_user(name):
    print("Hello", name)

greet_user("Zumi")
```

<font size=3> 我們透過改變變數(name)來改變函數打招呼的對象("Hello")。在這個例子中，"Zumi"就是那個變數，嘗試利用其他變數(name)呼叫這個函數，記得要使用引號引住變數！</font>

```python
greet_user("CHANGE ME!")
```

<font size=3>如果在函數中我們不使用引號或使用數字的話會怎樣？</font>

```python
greet_user(12345)
```

<font size=3> 由於我們定義了要在屏幕上打印(print)的功能，因此任何輸入都必須用引號引起來。但不代表你不能在函數中使用數字！Zumi有很多變數，例如：速度,持續時間等等都是數字。</font>
