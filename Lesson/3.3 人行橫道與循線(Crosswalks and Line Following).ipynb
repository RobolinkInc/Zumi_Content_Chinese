## Line Following(循線)
<font size=3> 現在把Zumi遇到線會停下來改為循著線行駛。在路上，Zumi會使用攝像頭檢測車道並保持在道中行駛。在這情況下，您需要使用一個函數來使Zumi循線行走，在學習編程前我們先了解它的原理：
    
<img src="../Data/images/line_follower.png" width=400>

我們將會使用底部的2個IR sensor來偵測線路。假裝您需跟隨地板面上的膠帶行走。如果您的右腳踏住膠帶，那您要向那邊轉向？您的左腳要怎麼移動？IR感應器亦是同樣原理。<br> <br>
    
<img src="../Data/images/black_line.png" width=400></font>

### Line_follower()
<font size=3> 利用導電膠紙或driving school mat(Zumi的配件)並將Zumi放到黑線的起點。<font face="Courier">zumi.line_follower()</font>需要至少一個變數:duration(持續時間)。舉個例，讓這個編程運行3秒。

```python
zumi.line_follower(3)
```

<font size=3> 如果您覺得Zumi跟線的接觸不良或光跟暗的差距不夠大的話，您可能需要改變**thresholds(界限)**或IR感應器所探測的數值範圍。默認為100。參考課節3.1以測試您的IR感應器並找出它的界限。</font><br>

<font size=3><font face="Courier">zumi.line_follower(3, left_thresh=100,right_thresh=100)</font>

<font size=3>**Note**: 這個函數包含一個while loop。如果雙個IR感應器都同時測到白色(線的終點)，編程便會自動停止，就算持續時間還沒結束。</font>
