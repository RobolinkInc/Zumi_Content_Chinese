<font size =3>還記得<font face="Courier">forward()</font>這個指令嗎? 它預設了Zumi將要40speed的速度向目前的方向行駛一秒，當然您也能更改這些數值(變數)。接下來所列出的函數均帶有默認的參數,如果您只利用名稱來呼叫出函數,例如:<font face="Courier">right_u_turn()</font>, Zumi便會以默認的參數做出一個u-turn。跟<font face="Courier">forward()</font>一樣，您亦可以將它們更改為所需的數值。</font>
