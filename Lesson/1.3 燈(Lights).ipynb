# Lights(燈)

<font size =3> 跟市面上的智能小車一樣，Zumi亦具備車頭大燈跟剎車燈給矛用家作編程使用。在電路板上，這些指示燈被稱為**LEDs(發光二極管)**. </font>

### Import libraries(導入庫)

````python
from zumi.zumi import Zumi
import time

zumi = Zumi()
````

## 什麼是LED?

<font size =3> LED 全名為 “light-emitting diode”.Diode(二極管)是一種電子設備，只允許電流在一個方向上流動。LEDs正在取代傳統的白熾燈泡，因為它們的效率更高。</font>

### Zumi的LEDs
<font size =3>Zumi板上有8個LED，但不是每一盞LED也能進行編程。Zuim的左面有三盞LED顯示充電狀態，而右面則有一盞白色的LED來顯示Pi成功啟動。Zumi的車頭大燈亦是LED，請注意不要跟紅外線感應器混淆。稍後您將會理解更多關於Zumi的車頭大燈，其實它們被外殼隱藏了所以不容易發現。 </font>

## Zumi LED Functions

<font size =3> 除了打開和關閉車頭大燈和剎車燈外，我們還可以使用轉向信號燈！
以下是可供您使用的所有功能：

```
* all_lights_on()

* all_lights_off()

* headlights_on()

* headlights_off()
       
* brake_lights_on()
       
* brake_lights_off()
        
* hazard_lights_on()
        
* hazard_lights_off()

* signal_left_on()
       
* signal_left_off()
       
* signal_right_on()
       
* signal_right_off()
```     

現在嘗試將轉向訊號合併到您的遙控(remote control)編程中或重新編寫一個開關燈的編程。
以下是編程範本，請預留空間供Zumi行駛:
</font>

````python
zumi.signal_right_on()
zumi.forward(duration=3)
zumi.turn_right()
zumi.signal_right_off()
zumi.forward()
````
