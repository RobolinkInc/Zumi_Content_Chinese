## go_straight()
<font size=3> 當您走路時，那些動作會不斷重複進行？右腳，左腳，右腳，左腳...Zumi 同樣也會以「一步」來向前行駛。當這些小步放進loop中，Zumi便會向前行駛。使Zumi以一小步行走的駕駛指令是 <font face="Courier"> zumi.go_straight()</font>。不同於<font face="Courier">forward()</font>, 參數(parameters)是不能任意改變的。Zumi要知道速度跟**航向(heading)**或方向。在以下的圖像中，Zumi的航向是什麼？
    
<img src="../Data/images/heading_straight.png" width="200">
    
運行以下的編程以觀察以<font face="Courier">go_straight()</font>行走的情況。

```python
zumi.reset_gyro() # Resetting all angles to zero at the start

for x in range(40): # Take 40 steps
    zumi.go_straight(40, -30) # Drive at speed 40 at heading -30
    
zumi.stop() # Don't forget to stop!
```

<font size=3>在結尾中的<font face="Courier">stop()</font>是非常重要的！由於 <font face="Courier">go_straight()</font>給予一個非常順暢且干穩的行駛，當中是無停頓的，同時亦沒有任何東西會停止Zumi繼續行走(除非電池用盡)。在loop完結後，編程便會進入下一組代碼。任何不屬於loop裡的代碼將不會縮排至loop中。<br><br>

## Conditionals(有條件的)
<font size=3> 自走小車通常都是以感應器來做出行駛決定的。接下來您將學習到如何編寫一個程式來將方向顛倒的Zumi停止行駛！</font>


### Orientation(方向)
<font size=3> 要知道Zumi是否方向/上下顛倒，您需要學習一個全新的函數: <font face="Courier">get_orientation()</font>。Zumi可能會面臨的六個狀態：
    
<img src="../Data/images/orientations.png" width="750">

使用下面的編程來驗證六個不同的狀態。如果您看到 0,-1,或7,Zumi是尚未進入狀態或進入了一個未知的狀態。


```python
for i in range(80):
    zumi.update_angles()
    orientation = zumi.get_orientation() # Save this value in a variable so that you can use it later...
    print(orientation)
    time.sleep(0.1)
    IPython.display.clear_output(wait=True) 
print("Done!")
```
