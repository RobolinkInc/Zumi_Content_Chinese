<font size =3>比賽分為三個獨立的挑戰。第一個涉及使用QR Code來告訴Zumi在哪裡停車。第二個要求zumi在停車標誌處停車，然後穿過移開停車標誌後的路段並在中途不接觸到任何牆壁。在第三次挑戰中，Zumi需停在紅牌處，並等待紅牌換成綠牌。之後，Zumi需巡線路行駛直至偵測到QR Code並讀取它，繼續沿著QR Code加密存儲的方向直到達終點。 </font>
