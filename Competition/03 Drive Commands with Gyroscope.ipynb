# Gyroscope(陀螺儀)

<font size =3>當您製作遙控器時，Zumi會根據您設定的度數(變數)向左或向右旋轉。您有思考過Zumi是怎樣知道它要轉多少度嗎？Zumi配備了一個**MPU(Memory protection unit/内存保護單元)**,或是motion processing unit(運動處理單元),以及具備兩個非常重要的感應器，它們是直線行駛及做出準確轉彎必需的配件: **gyroscope（陀螺儀)** and **accelerometer（加速規）**. 這節我們將先學習gyroscope(陀螺儀)，晚點再研究加速規！</font>


## What is a gyroscope?（什麼是陀螺儀？）
<font size =3> 當您想到陀螺儀時，您可能會想到一個固定在環上並會旋轉的車輪，無論您向哪個方向旋轉或翻轉陀螺儀，它都保持相同方向的平衡。為何會出現這個現象？這可以用動量守恆來解釋。您可能沒聽過這個原理，但在日常生活中其實很容易會接觸到。舉個例，如果您曾經在辦公椅(有車輪的)上轉過，可能會注意到屈膝把腳放到椅子並用手把住時旋轉得更快。由於必須保持動量，因此，當您使手臂和腿(質量)靠近您時，速度會加快。
<img src="../Data/images/gyroscope.jpg">

但在電子學裡陀螺儀並不是這樣的。它們是非常細小的芯川並根據動作來檢測方向的變化。. 在機械學中，陀螺儀則負責測量轉速。而在Zumi上，陀螺儀是非常重要的，因為我們它測量出的轉速來計算在特定時間上需要旋轉多少度。<br><br>
</font>

### Import libraries（導入庫）

```python
from zumi.zumi import Zumi
from zumi.util.screen import Screen
import time

zumi = Zumi()
screen = Screen()
```

## Axes(軸)
<font size =3> 您可以使用多個軸來測量轉速。最常使用的一個軸是偏擺軸(yaw),即是向左和向右轉彎。您還可以測量Zumi是向前傾斜還是向後傾斜或向左或向右傾斜。這三個軸稱為:roll(翻滾),pitch(俯仰),及yaw(偏擺).
    
<img src="../Data/images/zumi_axes_names.png" width=700>

以下有三組代碼可以讀取這三個軸: X, Y,及Z。運行以下的代碼,查看Zumi的屏幕來檢測角度如何受到影響。你能把X, Y,及 Z 配對成 roll(翻滾),pitch(俯仰),及yaw(偏擺)嗎?在運行編程前請確保Zumi在平坦的地面上行駛。</font>

### X-angle

```python
zumi.reset_gyro()
for i in range(0,50):
    current_angle = int(zumi.read_x_angle())
    message = " X-Angle reading           "
    message = message + str(current_angle)
    screen.draw_text(message)
    time.sleep(0.05)
    
print("Done")
screen.draw_text_center("Done")
```

### Y-angle

```python
zumi.reset_gyro()
for i in range(0,50):
    current_angle = int(zumi.read_y_angle())
    message = " Y-Angle reading           "
    message = message + str(current_angle)
    screen.draw_text(message)
    time.sleep(0.05)
    
print("Done")
screen.draw_text_center("Done")
```

### Z-angle


```python
zumi.reset_gyro()
for i in range(0,50):
    current_angle = int(zumi.read_z_angle())
    message = " Z-Angle reading           "
    message = message + str(current_angle)
    screen.draw_text(message)
    time.sleep(0.05)
    
print("Done")
screen.draw_text_center("Done")
```

### Reset Gyroscope(重設陀螺儀) 

<font size=3>您可能會注意到每個函數的開端都會出現<font face="Courier">reset_gyro()</font>這一組代碼.。這個函數會把roll(翻滾),pitch(俯仰),及yaw(偏擺)。事實上運行<font face="Courier">zumi=Zumi()</font>這個函數時便會同時這三個軸的數值了，但當您在編程的中後段時需要主行重置的話便需要使用這個函數了<font face="Courier">reset_gyro()</font>。 </font>

```python
zumi.reset_gyro() 
for i in range(100):
    z_angle = int(zumi.read_z_angle()) # <-- This function reads the angle
    message = " Z Angle reading        "
    message = message + str(z_angle)
    screen.draw_text(message)
    time.sleep(0.1)
```
