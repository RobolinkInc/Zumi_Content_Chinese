## Loops(循環)
<font size =3> 您可能已經註意到，您剛才重複了同一段代碼四次。這浪費不少時間在輸入上。為了解決這個問題，您可以使用loop(循環)，讓您重複部分代碼減化整個編程。 loop也是一個很好的shortcut(捷徑) --- 如果無有了它，您將總是不得不一遍又一遍地輸入相同的內容。 它之所以稱作為loops是因為完成最後一條語句便會從頭到尾再做一次。</font>

<img src="../Data/images/loop_square.png" width=300>

### For loops
<font size =3> 一個for loop 將根據編程設定的次數以重覆運行該次數。這意味著您可以自行選擇代碼重複多少次。以下是利用loop所編寫的正方形代碼(square code)，您發現到有什麼不同的地方嗎？<r><br>
    
<font face="Courier">
for x in range(4): <br>
<span style="margin-left: 40px;">zumi.forward()</span> <br>
<span style="margin-left: 40px;">zumi.turn_left()</span> <br>
</font>

 在這個例子中, x 是一個 **variable(變數)**,或稱為placeholder(佔位乎),用作檢查loops運行了多少次。您認為一個loop開始時的數字是多少？試運下以下編程並計算它的輸出出現了多少次。我們把每次的輸出都相隔了一秒，這更方便統計。觀察prints數值(x)的改變，每次遞進(**iteration**)1次。</font>


```python
for number in range(4):
    print(number)
    time.sleep(1)
```

<font size =3> 您有注意到變數(x/variable)由0開始並以3結束嗎？在Python的世界中， loops(循環)預設是從0開始並會以1來遞增。在編程的開端, x = 0。在編程的結尾, x=3。 變會將會停在3,因為loop只會在x少於括號中的數值時繼續運行。再增加1就會便成x=4。由於4不少於4，因此loop會中止。嘗試改變括號內的數值以改變遞增的次數，看看它將會停中什麼數字？

### Transformations(轉變) <br> <br>
<font size=3> 每一個編程的圖形都需要涉及原圖形的旋轉，反射或平移。<br> <br>
**Rotation** 表示圖形圍繞一個中心點旋轉。.<br>
**Reflection** 表示圖形在直線上翻轉。<br>
**Transformation** 表示圖形的每個點都沿相同方向移動相同距離。<br> <br>

在這三個情況中，圖形仍具有相同的大小，面積，角度和線長。
